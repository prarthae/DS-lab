```python
#superwised learning algorithm that classifies data
# it is used for both linear and nonlinear classification
# used for regression problems
# used for outlier ditection task
# applications : sentimental analysis,imag recognition and classification, hand writing identification, spam ditection, face detection, anomalie detection ,gene expression analysis etc.
# svm linear: is 2d serface a straight line that classifies data points,(also called maximum margin linear classifier-setting maximum margin)
# it ensures that the margin between the closest points(support vectors) of different classes is maximized
# svm non-linear : use hyper plane. dimaension of hyperplane depents on number of features. 
# In non linear svm there can be multiple planes or decission boundary to segrigate the classes in n diamensional space-we need to find out the best decission boundry that helps to classify the data points,
# this best boundary is known as the hyperplane, simply svm is a machine learnig algorithm to find the hyper plane in any n diamensional space that distictly classifies the data point
# whwn outliers are there need to miss classify or ignore it.
# if we consider outliers-minimum margin-wrong prediction
# soft margin formulation : it alows svm to make a certain number of mistakes and keep margin as wide as possibles so that other data points can still be classified correctly
# c-parameter : how to perform miss classification - with the help of hyper parameter called c-parameter, it tells the svm how much you want to avoid miss classifying in each training
# when u increase the value of c-parameter-minimum margin, decrease the value of c-parameter maximum margin
# errors in svm: 1. classification error(A), 2. Margin error(B)
# svm goal= min(cA+B)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report
import matplotlib.pyplot as plt
```

```python
X,y=make_classification(n_samples=1000,n_features=2,random_state=42,n_informative=2,n_redundant=0)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
model=SVC(kernel="linear")
```

```python
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
```

```python
accuracy_score(y_test,y_predict)
```

```python
report=classification_report(y_test,y_predict)
print(report)
```

```python
print(f"classification_report{report}")
```
