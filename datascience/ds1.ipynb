```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,11)
y=x+5

plt.title("graphical representation")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,'--')
plt.show()
```

```python
plt.title("graphical representation")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,'*r',)
plt.show()
```

```python
plt.title("graphical representation")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,'om')
plt.show()
# m - magenta, k- black
```

```python

y=3*x+5
plt.title("graphical representation")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,color='orange',marker='o',markersize='15',label='graph')
plt.legend()
plt.show()
```

```python
x=[1992,1995,1998,2001,2004]
y=[100,159,159,300,250]
z=[10,26,58,110,126]

plt.title("graphical representation")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,color='green',marker='v',markersize='9',label='graph')
plt.plot(x,z,color='red',marker='*',label='others')
plt.legend()
plt.show()
```

```python
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)

plt.title("Sin function")
plt.xlabel("value")
plt.ylabel("sin(x)")
plt.plot(x,y,color='violet',label='graph')
plt.legend()
plt.show()
```

```python
plt.title("Cos function")
plt.xlabel("value")
plt.ylabel("cos(x)")
plt.plot(x,z,color='red',label='graph')
plt.legend()
plt.show()
```

```python
plt.title("Sin function")
plt.xlabel("value")
plt.ylabel("sin(x)")
plt.plot(x,y,color='red',label='sin',marker='v',markersize='2')
plt.plot(x,z,color='green',label='cos',marker='*',markersize='2')
plt.legend()
plt.show()
```

```python
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y,color='red')
plt.title("sine function")
plt.subplot(2,1,2)
plt.plot(x,z)
plt.title("cos function")
plt.show()
```

```python
x=[45,90,30,5]
label=['p','q','r','s']
plt.legend()
plt.pie(x,labels=label)
plt.legend()
plt.show()
```

Q1. write a python code plot aline chart to depict the run rate of t-20 match from the given data
    overs=5,10,15,20
    runs=45,79,145,234
    also change the thickness of the line

```python
over=[5,10,15,20]
run=[45,79,145,243]

plt.title("T-20 Match")
plt.xlabel("overs")
plt.ylabel("runs")
plt.plot(over,run,linewidth=3,color='red',label='rate')
plt.show()
```
