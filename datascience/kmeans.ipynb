```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs #generate sample data
```

```python
data,_=make_blobs(n_samples=300)
print(data)
```

```python
kmeans=KMeans(n_clusters=5,random_state=42,n_init="auto").fit(data)#initialization of centroid n_init="auto"
labels=kmeans.labels_ #it contains the cluster (label), specifying which cluster each data points belongs to
set(labels)
```

```python
centroids=kmeans.cluster_centers_ #stores the coordinates of the centroids for these five clusters
```

```python
plt.title("Scatter plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.scatter(data[:,0],data[:,1],c=labels,cmap="viridis",marker="o") ##this will plot the data points,and color in each point accr to their labels 
plt.show()
```

```python
plt.title("Scatter plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.scatter(data[:,0],data[:,1],c=labels,cmap="viridis",marker="o") ##this will plot the data points,and color in each point accr to their labels 
plt.scatter(centroids[:,0],centroids[:,1],marker="*",color="red")
plt.legend(list(labels))
plt.show()
```
